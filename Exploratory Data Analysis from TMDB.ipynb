```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

# Loading the Data

```python
df_2000 = pd.read_csv('/Users/christianrim/Documents/final_tmdb_data_2000.csv',lineterminator='\n')
df_2001 = pd.read_csv('/Users/christianrim/Documents/final_tmdb_data_2001.csv',lineterminator='\n')
```

```python
# display the 2000 movie dataset
df_2000.head()
df_2001.info()
```

```python
df =  pd.concat([df_2000,df_2001])
df.head()
```

```python
df.info()
```

# Cleaning the data

```python
# Removing columns with a lot of missing values
df = df.drop(['belongs_to_collection','backdrop_path','poster_path','tagline','homepage'], axis=1)
df.head()
```

```python
df['budget'].describe()
```

```python
# getting rid of the 0's in the 'budget' and 'revenue' column
filtered_df = df[(df['budget'] != 0) & (df['revenue'] != 0)]
```

```python
filtered_df.head()
```

## How many movies had at least some valid financial information 

```python
# Select desired columns
selected_cols = ['title', 'revenue', 'budget']
result_df = filtered_df.loc[:, selected_cols]
result_df[:10]
```

```python
df_filtered = df[(df['revenue'] > 0) & (df['budget'] > 0)]
# Group the filtered DataFrame by 'movie' and get the count of movies in each group
count_by_movie = df_filtered.groupby('title').size()

# Display the result
print(count_by_movie)
```

- There were 402 movies with at least one financial data. 

## How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# counting the movie count for each certification category 
grouped_movies = filtered_df.groupby('certification')
count_movies = grouped_movies.size() 
print(count_movies) 
print(count_movies.sum())
```

- There are 318 movies for each certification categories. 

## What is the average revenue per certification category?

```python
# Set the display float format option to a non-scientific notation format
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Group the data by "certification" and calculate the average revenue for each group
avg_revenue_by_certification = filtered_df.groupby("certification")["revenue"].mean()
avg_revenue_by_certification
```

## What is the average budget per certification category?

```python
# Set the display float format option to a non-scientific notation format
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Group the data by "certification" and calculate the average revenue for each group
avg_budget_by_certification = filtered_df.groupby("certification")["budget"].mean()
avg_budget_by_certification
```
